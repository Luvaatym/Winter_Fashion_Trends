```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

```

```python
df = pd.read_csv('Winter_Fashion_Trends_Dataset.csv')
```

```python
df['Price(USD)'] = df['Price(USD)'].fillna(df['Price(USD)'].median())
```

```python
df['Category'] = df['Category'].fillna(df['Category'].mode()[0])
df['Brand'] = df['Brand'].fillna(df['Brand'].mode()[0])
```

```python
df.duplicated().sum()
```

```python
df = df.drop_duplicates()
```

```python
from scipy import stats

numeric_df = df[['Price(USD)', 'Popularity_Score', 'Customer_Rating']]
z_scores = np.abs(stats.zscore(numeric_df))
outliers = (z_scores > 3).any(axis=1)

df[outliers]
```

```python
plt.figure(figsize=(6,4))
sns.boxplot(x=df['Price(USD)'])
plt.title('Boxplot ')
plt.show()
```

```python
plt.figure(figsize=(8,4))
sns.boxplot(x='Category', y='Price(USD)', data=df)
plt.title('   ')
plt.xticks(rotation=45)
plt.show()
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title(' ')
plt.show()
```
